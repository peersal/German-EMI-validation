For this script to work the wanted Folders of this website https://opendiscourse.de/ have to be in the same directory.

```python
import pandas as pd

# Specify the chunk size
chunksize = 100000  # Adjust this number based on your memory constraints

# Initialize an empty DataFrame to store results if needed
result = pd.DataFrame()

df = pd.DataFrame()

progress = 0

# Process the file in chunks
for chunk in pd.read_csv('dataverse_files/csv/speeches.csv', chunksize=chunksize):
    progress +=1
    print(f"{progress} Chunks loaded")
    
    # Filter chunks where the year part of the 'date' column is >= 2023
    filtered_chunk = chunk[chunk['positionShort'].isin(['Member of Parliament', 'minister'])]

    
    df = pd.concat([df, filtered_chunk])
```

```python
df = df.reset_index(drop=True)
if 'level_0' in df.columns:
    df = df.drop(columns=['level_0'])

```
